```python
missingdata=df.isnull().mean() * 100
missingdata
```

```python
def visualize_data(data, caption='', ylabel='percentage of missing data'):
    sb.set(rc={'figure.figsize':(15, 8.27)})
    plt.xticks(rotation=90)
    
    fig = sb.barplot(x = data.keys()[:min(40 , len(data))].tolist() , y = data.values[: min(40 , len(data))].tolist()) \
    .set_title(caption)
    
    # set labels
    plt.ylabel(ylabel)
    
    plt.show()
```

```python
visualize_data(missingdata, 'percentage of missing data')
```

```python
df.Acronym=df.Acronym.fillna('missing Acronym')
```

```python
df.isnull().mean() * 100
```

```python
categorical_features=df.select_dtypes(include=object).columns

features=categorical_features[df[categorical_features].isnull().mean() > 0]

for feature in features:
    df[feature]=df[feature].fillna(f'missing{features}')
```
